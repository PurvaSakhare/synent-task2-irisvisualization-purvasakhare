```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

iris = pd.read_csv('Iris.csv')
```

```python
sns.histplot(iris['SepalLengthCm'])
plt.show()
```

```python
sns.scatterplot(
x='SepalLengthCm',
y='PetalLengthCm',
hue='Species',
data=iris
)
plt.show()
```

```python
iris['Species'].value_counts().plot(kind='bar')
plt.show()
```
